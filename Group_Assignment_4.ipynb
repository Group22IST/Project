```python
import sqlite3
connectionToDatabase = sqlite3.connect('UberData1')
cursor = connectionToDatabase.cursor()
```

```python
cursor.execute('''Create table IF NOT EXISTS TripData(TripID text PRIMARY KEY, Distance text, Location text, Start Time text, WaitTime text, CustomerRating text, DriverRating text, DriverID text, CustomerID text, TransactionID text, CardID text)''')
connectionToDatabase.commit()
```

```python
cursor.execute('''Create table IF NOT EXISTS DriverData(DriverID text PRIMARY KEY, First Name text, Last Name text, DriverRating text, CarID text)''')
connectionToDatabase.commit()
```

```python
cursor.execute('''Create table IF NOT EXISTS CarData(CarID text PRIMARY KEY, Latitude text, Longitude text, License Plate text, UberType text, DriverID text)''')
connectionToDatabase.commit()
```

```python
cursor.execute('''Create table IF NOT EXISTS TransactionData(TransactionID text PRIMARY KEY, Paymenttype text, Amount text, CustomerID text, TripID text, CarID text, DriverID text)''')
connectionToDatabase.commit()
```

```python
cursor.execute('''Create table IF NOT EXISTS CustomerData(CustomerID text PRIMARY KEY, First Name text, Last Name text, PreviousTripID text)''')
connectionToDatabase.commit()
```

```python
cursor.execute('''Create table IF NOT EXISTS PreviousTripData(PreviousTripID text PRIMARY KEY, trip1 text, trip2 text, trip3 text)''')
connectionToDatabase.commit()
```
